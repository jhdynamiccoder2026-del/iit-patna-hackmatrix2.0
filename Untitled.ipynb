## Data Extraction Using "PyMUPDF"
### In this only Electronically prepared document will come

```python
def extract_text_for_api(pdf_path):
    # 1. Open the document
    doc = fitz.open(pdf_path)
    
    # 2. Extract all text into one giant string
    full_text = ""
    for page in doc:
        full_text += page.get_text() + "\n"
        
    doc.close()
    
    # 3. Create the exact dictionary format your FastAPI expects
    api_payload = {
        "text": full_text.strip()
    }
    
    # 4. Convert to JSON string
    json_result = json.dumps(api_payload, indent=4, ensure_ascii=False)
    
    return json_result

# --- Example usage ---
pdf_file = r"C:\Users\gauta\Downloads\Copy of Copy of report26.pdf"
extracted_json = extract_text_for_api(pdf_file)

print(extracted_json)
```

# Integration of NLP using API

```python
from fastapi import FastAPI
from pydantic import BaseModel
from fastapi.middleware.cors import CORSMiddleware
import spacy
import re

app = FastAPI(title="Smart Clinical Extraction API")

app.add_middleware(
    CORSMiddleware,
    allow_origins=["*"],
    allow_credentials=True,
    allow_methods=["*"],
    allow_headers=["*"],
)

print("Loading NLP models... Please wait.")
nlp_general = spacy.load("en_core_web_sm")
nlp_medical = spacy.load("en_ner_bc5cdr_md")
print("Models loaded successfully!")

class ClinicalTextRequest(BaseModel):
    text: str

# ==========================================
# 1. THE AI ENGINE (For Unstructured Prose)
# ==========================================
def extract_with_nlp(text: str):
    doc_general = nlp_general(text)
    doc_medical = nlp_medical(text)
    extracted = {"Name": None, "Age": None, "Diseases": [], "Medications": []}

    for ent in doc_general.ents:
        if ent.label_ == "PERSON" and not extracted["Name"]:
            extracted["Name"] = ent.text
        elif ent.label_ == "DATE" and "year" in ent.text.lower() and not extracted["Age"]:
            extracted["Age"] = ent.text

    for ent in doc_medical.ents:
        if ent.label_ == "DISEASE":
            extracted["Diseases"].append(ent.text)
        elif ent.label_ == "CHEMICAL":
            extracted["Medications"].append(ent.text)

    extracted["Diseases"] = list(set(extracted["Diseases"]))
    extracted["Medications"] = list(set(extracted["Medications"]))
    return extracted

# ==========================================
# 2. DISCHARGE SUMMARY ENGINE
# ==========================================
def extract_discharge_summary(text: str):
    extracted = {"Name": None, "Age": None, "Diseases": [], "Medications": []}
    
    name_match = re.search(r"Name:\s*([A-Za-z\s]+)", text)
    if name_match: extracted["Name"] = name_match.group(1).strip()

    age_match = re.search(r"Age/Gender:\s*(\d+)", text)
    if age_match: extracted["Age"] = age_match.group(1).strip()

    diagnosis_match = re.search(r"FINAL DIAGNOSIS:\s*\n\s*(.+)", text)
    if diagnosis_match: extracted["Diseases"].append(diagnosis_match.group(1).strip())

    meds_block_match = re.search(r"ADVICE ON DISCHARGE:(.*?)FOLLOW UP ON", text, re.DOTALL)
    if meds_block_match:
        med_lines = re.findall(r"\d+\.\s*(.+?)(?:;|ORAL|SYRUP)", meds_block_match.group(1))
        for med in med_lines:
            if med.strip(): extracted["Medications"].append(med.strip())

    return extracted

# ==========================================
# 3. LAB & PATHOLOGY REPORT ENGINE
# ==========================================
def extract_lab_report(text: str):
    extracted = {"Name": None, "Age": None, "Diseases": [], "Medications": []}
    
    # Matches "Patient Name \n : anonymous"
    name_match = re.search(r"Patient Name\s*:\s*([^\n]+)", text)
    if name_match: extracted["Name"] = name_match.group(1).strip()

    # Matches "Age/Sex \n : 25 Yr/F"
    age_match = re.search(r"Age/Sex\s*:\s*([^\n]+)", text)
    if age_match: extracted["Age"] = age_match.group(1).strip()

    # Look for Interpretation or Impression
    impression_match = re.search(r"(?:IMPRESSION|INTERPRETATION\s*/RESULT):\s*(.*?)(?=\nNote|\n\*|\n\n)", text, re.IGNORECASE | re.DOTALL)
    if impression_match: 
        # Clean up the extracted diagnosis block
        clean_disease = impression_match.group(1).replace("\n", " ").strip()
        extracted["Diseases"].append(clean_disease)

    return extracted

# ==========================================
# 4. THE SMART ROUTER ENDPOINT
# ==========================================
@app.post("/extract")
async def extract_details(request: ClinicalTextRequest):
    text = request.text
    
    # Analyze the document type
    if "Discharge Summary" in text:
        engine_used = "Regex (Discharge Summary)"
        results = extract_discharge_summary(text)
    elif "Cytology" in text or "Department of Pathology" in text or "Sample Type" in text:
        engine_used = "Regex (Lab/Pathology Report)"
        results = extract_lab_report(text)
    else:
        engine_used = "NLP (Unstructured Prose)"
        results = extract_with_nlp(text)

    return {
        "status": "success", 
        "engine_used": engine_used,
        "data": results
    }
```
